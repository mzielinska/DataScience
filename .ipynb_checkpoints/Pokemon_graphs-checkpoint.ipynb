```python
# Add pokemon names
df = pd.merge(df, pnames, how='left', on = "pokemonId")
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# What are the most popular pokemon types?
ptype = df['pokemon_name'].value_counts()[:35].sort_values()

ptype.plot(kind ='barh',)
plt.show()
```

```python
# Most popular cities where pokemons appeared
pcity = df['city'].value_counts()[:30].sort_values()
pcity.plot(kind ='barh',)
plt.show()
```

```python
df['n'] = 1
```

```python
table = df.pivot_table(values = 'n', index = 'city', columns = 'pokemon_name', aggfunc =np.sum)
print(table)
```

```python
table = pd.pivot_table(df, values = 'n' , index = 'city', columns = 'pokemon_name', aggfunc = 'count')
ax = sns.heatmap(table)
```

```python
df_europe = df[df['continent']=='Europe']

table = pd.pivot_table(df_europe, values = 'n' , index = 'city', columns = 'pokemon_name', aggfunc = 'count')
ax = sns.heatmap(table, cbar =False)
```

```python
table = pd.pivot_table(df_europe, values = 'n' , index = 'appearedDay', columns = 'appearedMonth', aggfunc = 'count')
ax = sns.heatmap(table, cbar =False)
```

```python
pcontinent = df['continent'].value_counts().sort_values()
pcontinent.plot(kind ='barh',)
```

```python
date = df['appearedDay'] + '/' + df['appearedMonth'] + '/' + df['appearedYear']
df['date'] = date
time = df['appearedHour'] + ':' + df['appearedMinute']
df['time'] = time
```
